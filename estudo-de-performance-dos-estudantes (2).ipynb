- Fonte de dados

> https://archive.ics.uci.edu/dataset/320/student+performance

- Objetivo

> Analisar a interferência de fatores na nota final de um aluno.

- Atributo-Alvo

>  Nota final do período (0 - 20)

- Atributos preditivos

> Sexo, idade, moradia, entre outros.

- Tipo de tarefa

> **Regressão**



# Compreensão De Dados

- Registros e Atributos

> São 649 registros (alunos) com 33 atributos, sendo 1 atributo-alvo e 32 preditivas.

- Tipos de variáveis

> Nesse dataset existem variáveis categoricas(16) e variáveis numericas(17). Não existe nemhuma variável com valor ausente 

- Inconsistências

> A principal das inconsistências acontece exatamente no atributo-alvo, quando o aluno tem condicões muito favorãveis a nota pode-se passar de 20 ou se as condições forem muito desfavoráveis as notas podem ser abaixo de 0.

- Distribuiçao do Atributo-Alvo

> As notas se concentram entre 10 e 15 e quase não existem alunos que alcançaram 20 ou 0.

# Análise exploratoria

- Relações entre atributo e o alvo.

> Quando a relação é positiva, isso quer dizer que os valores sao diretamente proporcionais, ou seja, se um aumentar o outro aumenta, por exemplo, o tempo de estudo, quanto mais tempo passar estudando a nota será maior, já a quantidade de faltas é negativa e inversamente proporcional, quanto maior o número de faltas, menor será a nota.

# Tabela de dispersão (studytime x G3)

> É perceptivel que a media de G3 aumenta conforme a quantidade de horas de estudo semanal

# Modelagem, avaliação e discurssão

-> *G1 E G2 são as notas anteriores*  



- Modelos Utilizados

> Regressão linear, Árvore de decisão e Random Forest.

- Principais paramêtros

> Número de faltas e reprovações anteriores.

- Comparação de resultados da modelagem

> O melhor modelo dos que foram testados foi o Random Forest, com MAE, RMSE e MSE médios (soma dos cenários com e sem G1/G2) de 2,12, 5,82 e 1,09 respectivamente.

- Comparaçao de resulatos das métricas

>  Podemos perceber que quando retiramos G1 e G2 da análise o MAE E RMSE aumentam, ja o MSE diminui, complica a análise apenas com fatores sócio-demográficos.

- Limitações

> G1 e G2 tem forte peso na interpretação dos dados, porém os dados sócio-demográficos dos alunos podem ter mudado desse as provas anteriores.

- O que poderia ser melhorado

> Ter um atributo categórico que definisse a dificuldade de cada prova, para que G1 e G2 nao tivessem tanta interferência no meio de tantas outros atributos.
